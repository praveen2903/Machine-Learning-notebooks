# Reinforcement Learning

Reinforcement learning in nutshell is to program an agent to be intelligent. It does so by taking actions and maximizing the rewards. 

### Agent
The thing that senses the environment. Thing we're trying to code intelligence into

### Environment
Real or simulated world the agent lives in

### State
Different configuration the agent can get into

### Rewards
Based on the environment, the agent, depending on the move, will either get a reward or punishment. The agent will not only try to maximize the immediate rewards, but future rewards as well. Also, reward is always a real number.

### Actions
What agent does to the environment. Eg: go up, go down, go left, go right.

### Episodes
An episode represents one run of the game. We may need to simulate the game thousands of times (episodes) before our RL agent learns. The number of episodes is a hyperparameter. 

### Terminal states
A terminal state is where a game ends and no more action can be taken. 



Every game is a sequence of states, actions and rewards. Convention is to start in state $S(t)$ take Action $A(t)$ and receive rewards $R(t)$. The action also brings the state to $S(t + 1)$

## Exploration vs Exploitation dilemma

This is a prominent problem in RL: whether you should explore, or exploit. That is, whether you should be choosing the best decision, or take a random decision to see if there's a better reward possible. If we always take the best decision (exploit) then we might miss the states with hidden rewards. 

There are many ways to solve this problem:

1. Using Epsilon

We take a random decision (explore) even though we know that this might not be the best decision. 

Pseudo code looks like:
```
if random() < eps:
    take_random_action()
else:
    take_best_action()
```
Where $\epsilon$ is the exploration probability which is usually around 5-10%. Also, in practise it is good to add a decay parameter which shrinks the value after a while since it is not wise to randomize choices when you are pretty sure which ones give better rewards. 

2. Using Epsilon with Optimistic initial values

In the previous method we initialized the means of all the options to 0. So if we choose the first option and have a mean $x$ then all the other options will have 0 mean, so the best agent becomes the one which we choose in the beginning. To avoid this, we initialize all the options with a high value so all of them are explored quickly. 

3. Upper Confidence Bound:

If there are two options: 1 with a mean of 100 and number of times = 10, and other = mean of 80 but the number of times the option was selected = 100 then we should go with the second option because the rewards have been averaged over more number of trials. UCB takes into account this info and makes choice accordingly. Here the reward is not just the mean but the "upper confidence bound" of that option. The epsilon is modified to a different values which is not static and changes over time. 

4. Thompson Sampling

Model the underlying distribution using betavariate distribution. It will get more accurate as the number of trials increase. 

## Notation

RL systems are represented by notations. Specifically this tuple: (s, a, r, s') where s is the current state, a is the action taken, r is the reward for action a, and s' is the state achieved by taking the action a. s' doesn't necessarily mean state at time t + 1. 

## Note
Always design the rewards in such a way that the agent comes up with how to acheive it, and not you. For instance, if you gave it rewards on intermediate states, then it may or may not win this game and just end up taking many intermediate rewards. Hence, give it rewards only when it wins the game.
